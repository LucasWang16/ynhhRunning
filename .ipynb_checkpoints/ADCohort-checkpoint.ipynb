```python
## pip3 install pyarrow fastparquet matplotlib seaborn
## Go to Kernel --> Restart Kernel after installing the libraries

import json
import pandas as pd
import os
import copy
import duckdb
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import hashlib
```

```python
## Configure the path for use in the sql-based query
path = '/home/jupyter/2835794-data'
```

```python
## AD cohort query
## Pulls ICD-10 G30.* (core Alzheimer's disease)
## Plus ICD-9 331.0 for backward compatibility if any legacy data exists
## NOTE: Not filtering by condition_type_concept_id here — pull broad, filter later
##       (Schizoph notebook used 32020 = EHR encounter dx; AD often appears on problem lists too)

query = f"""
SELECT
  per.person_id
  ,per.birth_datetime AS BIRTH_DATE
  ,cond.condition_source_value AS ICD
  ,cond.condition_start_datetime AS START_DATE
  ,cond.condition_type_concept_id AS COND_TYPE
FROM '{path}/person/*.parquet' per
  INNER JOIN '{path}/condition_occurrence/*.parquet' cond
    ON per.person_id = cond.person_id
    AND (
         cond.condition_source_value LIKE 'G30%'      -- ICD-10 Alzheimer's disease (all subtypes)
      OR cond.condition_source_value LIKE 'G30.0%'    -- early onset
      OR cond.condition_source_value LIKE 'G30.1%'    -- late onset
      OR cond.condition_source_value LIKE 'G30.8%'    -- other AD
      OR cond.condition_source_value LIKE 'G30.9%'    -- AD unspecified
      OR cond.condition_source_value = '331.0'        -- ICD-9 Alzheimer's (legacy)
    )
"""
## Note: 'G30%' already covers G30.0/G30.1/G30.8/G30.9 — the subtype LIKEs are
## redundant but kept explicit for readability / easy commenting-out.
```

```python
## Run query, materialize to pandas
df_ad_raw = duckdb.sql(query).to_df()
df_ad_raw.shape
```

```python
## Quick sanity checks before deciding the cohort definition

# 1. ICD code distribution — which subtypes show up, and at what counts
df_ad_raw['ICD'].value_counts()
```

```python
# 2. condition_type_concept_id distribution
# (32020 = EHR encounter dx, 32840 = EHR problem list, 32817 = EHR claim, etc.)
df_ad_raw['COND_TYPE'].value_counts()
```

```python
# 3. Per-person event counts
events_per_person = df_ad_raw.groupby('person_id').size()
print(f"Unique persons with any AD code: {df_ad_raw['person_id'].nunique()}")
print(f"Median AD codes per person: {events_per_person.median()}")
print(f"Persons with >=2 AD codes: {(events_per_person >= 2).sum()}")
```

```python
# 4. Age at first AD code (sanity: should mostly be >=60, with a tail for early-onset)
df_first = (df_ad_raw
            .sort_values(['person_id', 'START_DATE'])
            .groupby('person_id')
            .first()
            .reset_index())
df_first['AGE_AT_FIRST_AD'] = (
    (pd.to_datetime(df_first['START_DATE']) - pd.to_datetime(df_first['BIRTH_DATE']))
    .dt.days / 365.25
)
df_first['AGE_AT_FIRST_AD'].describe()
```

```python
# 5. Visualize age distribution
plt.figure(figsize=(8, 4))
sns.histplot(df_first['AGE_AT_FIRST_AD'].dropna(), bins=50)
plt.xlabel('Age at first AD code (years)')
plt.ylabel('Count')
plt.title(f'AD cohort: age at first diagnosis (n={len(df_first)})')
plt.axvline(65, color='red', linestyle='--', label='65 (early/late onset cutoff)')
plt.legend()
plt.show()
```

```python
## Define the analytic cohort
## Two common definitions — pick based on your research aim:

## (A) Liberal: any single G30.* code ever
cohort_liberal = df_ad_raw['person_id'].unique()
print(f"Liberal cohort (>=1 AD code): {len(cohort_liberal)}")

## (B) Strict: >=2 AD codes on separate days, >=30 days apart
##     (standard practice in AD EHR phenotyping — reduces rule-out / placeholder codes)
def strict_ad_cohort(df, min_codes=2, min_gap_days=30):
    df2 = df.copy()
    df2['START_DATE'] = pd.to_datetime(df2['START_DATE'])
    g = df2.groupby('person_id')['START_DATE'].agg(['min', 'max', 'nunique'])
    g['span_days'] = (g['max'] - g['min']).dt.days
    keep = g[(g['nunique'] >= min_codes) & (g['span_days'] >= min_gap_days)].index
    return keep.tolist()

cohort_strict = strict_ad_cohort(df_ad_raw)
print(f"Strict cohort (>=2 codes, >=30 days apart): {len(cohort_strict)}")
```

```python
## Build the final cohort table: one row per person, with index date = first AD code

df_cohort = (df_ad_raw[df_ad_raw['person_id'].isin(cohort_strict)]
             .sort_values(['person_id', 'START_DATE'])
             .groupby('person_id')
             .agg(
                 BIRTH_DATE=('BIRTH_DATE', 'first'),
                 FIRST_AD_DATE=('START_DATE', 'min'),
                 LAST_AD_DATE=('START_DATE', 'max'),
                 N_AD_CODES=('ICD', 'count'),
                 FIRST_AD_ICD=('ICD', 'first'),
             )
             .reset_index())

df_cohort['AGE_AT_FIRST_AD'] = (
    (pd.to_datetime(df_cohort['FIRST_AD_DATE']) - pd.to_datetime(df_cohort['BIRTH_DATE']))
    .dt.days / 365.25
)

df_cohort.head()
```

```python
## Save cohort for downstream use
df_cohort.to_parquet('ad_cohort.parquet', index=False)
print(f"Saved {len(df_cohort)} AD patients to ad_cohort.parquet")
```

```python
# 6. Investigate the suspicious young cases
young_cases = df_first[df_first['AGE_AT_FIRST_AD'] < 40].sort_values('AGE_AT_FIRST_AD')
print(f"Persons with first AD code before age 40: {len(young_cases)}")
young_cases[['person_id', 'BIRTH_DATE', 'START_DATE', 'ICD', 'AGE_AT_FIRST_AD']].head(20)
```

```python
# 7. Apply cohort filters
# Standard AD EHR phenotype:
#   - >=2 AD codes (already 29,275 candidates)
#   - >=30 days apart (filters out cluster-coded single episodes)
#   - age at first code >= 40 (filters out implausible cases)

MIN_AGE = 40
MIN_CODES = 2
MIN_GAP_DAYS = 30

df_ad_raw['START_DATE'] = pd.to_datetime(df_ad_raw['START_DATE'])
df_ad_raw['BIRTH_DATE'] = pd.to_datetime(df_ad_raw['BIRTH_DATE'])

g = df_ad_raw.groupby('person_id').agg(
    first_date=('START_DATE', 'min'),
    last_date=('START_DATE', 'max'),
    n_unique_days=('START_DATE', 'nunique'),
    birth_date=('BIRTH_DATE', 'first'),
)
g['age_at_first'] = (g['first_date'] - g['birth_date']).dt.days / 365.25
g['span_days'] = (g['last_date'] - g['first_date']).dt.days

cohort_strict = g[
    (g['n_unique_days'] >= MIN_CODES) &
    (g['span_days'] >= MIN_GAP_DAYS) &
    (g['age_at_first'] >= MIN_AGE)
].index.tolist()

print(f"Final strict AD cohort: {len(cohort_strict)}")
print(f"  Excluded by <2 codes:        {(g['n_unique_days'] < MIN_CODES).sum()}")
print(f"  Excluded by <30 day span:    {((g['n_unique_days'] >= MIN_CODES) & (g['span_days'] < MIN_GAP_DAYS)).sum()}")
print(f"  Excluded by age <40:         {((g['n_unique_days'] >= MIN_CODES) & (g['span_days'] >= MIN_GAP_DAYS) & (g['age_at_first'] < MIN_AGE)).sum()}")
```

```python
# 8. Build the final cohort table — one row per patient, index_date = first AD code
df_cohort = (df_ad_raw[df_ad_raw['person_id'].isin(cohort_strict)]
             .sort_values(['person_id', 'START_DATE'])
             .groupby('person_id')
             .agg(
                 BIRTH_DATE=('BIRTH_DATE', 'first'),
                 INDEX_DATE=('START_DATE', 'min'),       # first AD code = index date
                 LAST_AD_DATE=('START_DATE', 'max'),
                 N_AD_CODES=('ICD', 'count'),
                 N_AD_DAYS=('START_DATE', 'nunique'),
                 FIRST_AD_ICD=('ICD', 'first'),
                 EVER_G30_0=('ICD', lambda x: (x.str.startswith('G30.0')).any()),
                 EVER_G30_1=('ICD', lambda x: (x.str.startswith('G30.1')).any()),
                 EVER_G30_8=('ICD', lambda x: (x.str.startswith('G30.8')).any()),
                 EVER_G30_9=('ICD', lambda x: (x.str.startswith('G30.9')).any()),
             )
             .reset_index())

df_cohort['AGE_AT_INDEX'] = (
    (df_cohort['INDEX_DATE'] - df_cohort['BIRTH_DATE']).dt.days / 365.25
)

# Sanity peek
print(df_cohort.shape)
df_cohort.head()
```
